```python
from  selenium import webdriver
browser = webdriver.Chrome('chromedriver.exe')
```

```python
import bs4
import pandas as pd
```

```python
url = "https://www.imdb.com/chart/top?ref_=nv_mv_250"
browser.get(url)
```

```python
subbb=[]
```

```python
len(subbb)
```

```python
import time
sub_urls = browser.find_element_by_css_selector('tbody.lister-list').find_elements_by_css_selector('td.titleColumn')
try:
    for subs in sub_urls:
        sub_url = subs.find_element_by_css_selector('a').get_attribute('href')
        subbb.append(sub_url)
except:
    pass
```

```python
imdb_top250=pd.DataFrame()
for sub in subbb:
    browser.get(sub)
    time.sleep(3)
    parent = browser.find_element_by_css_selector('div[id = "wrapper"]')
    
    
    try:
        title_block = parent.find_element_by_css_selector('div.title_block')
        title_bar = title_block.find_element_by_css_selector('div.titleBar')
        title_wrapper = title_bar.find_element_by_css_selector('div.title_wrapper')
        title_wrapper_hi = title_bar.find_element_by_css_selector('div.title_wrapper').find_element_by_css_selector('h1')
        #Getting the Movie Name
        movie_name = title_wrapper_hi.get_attribute('innerHTML')
        mov_name = movie_name[0:movie_name.find('&')]
    except:
        mov_name=float('nan')
    #Getting the Year  
    try:
        year =title_wrapper_hi.find_element_by_css_selector('span[id="titleYear"]').find_element_by_css_selector('a').get_attribute('innerHTML')
    except:
        year=float('nan')
    #Getting thr IMDB_Rating
    try:
        IMD_Rating = title_block.find_element_by_css_selector('span[itemprop="ratingValue"]').get_attribute('innerHTML')
    except:
        IMD_Rating=float('nan')
    #Getting No of reviewers
    try:
        No_of_reviewers = int(title_block.find_element_by_css_selector('span[itemprop="ratingCount"]').get_attribute('innerHTML').replace(',',''))
    except:
        No_of_reviewers=float('nan')
    #Getting censor board rating details
    try:
        censor_board= title_wrapper.find_element_by_css_selector('div.subtext').get_attribute('innerHTML').replace(' ','').replace('\n','')
        censor_board_rating=censor_board[0:censor_board.find('<spanclass="ghost">')]

    except:
        censor_board_rating=float('nan')
    #Getting the runtime of a movie
    try:
        run_time =title_wrapper.find_element_by_css_selector('time').get_attribute('innerHTML').replace(' ','').replace('\n','')
    except:
        run_time=float('nan')
    #Getting the genre
    try:
        genres = title_wrapper.find_element_by_css_selector('div.subtext').find_elements_by_css_selector('a')
        gens = genres[0:len(genres)-1]
    except:
        pass

    if len(gens):
        genre1 = gens[0].get_attribute('innerHTML')
        gens.remove(gens[0])
    else:
        genre1=float('nan')
    if len(gens):
        genre2 = gens[0].get_attribute('innerHTML')
        gens.remove(gens[0])
    else:
        genre2=float('nan')
    if len(gens):
        genre3 = gens[0].get_attribute('innerHTML')
        gens.remove(gens[0])
    else:
        genre3=float('nan')
    if len(gens):
        genre4 = gens[0].get_attribute('innerHTML')
        gens.remove(gens[0])
    else:
        genre4=float('nan')
    #Getting the release date of the movie
    try:
        release_date =genres[len(genres)-1].get_attribute('innerHTML')
    except:
        release_date=flaot('nan')
    #Getting the summary of a movie
    try:
        summary =parent.find_element_by_css_selector('div.summary_text').get_attribute('innerHTML').replace('\n','').replace('  ','')
    except:
        summary=flaot('nan')
    #Getting the writers of a movie
    try:
        credits = parent.find_elements_by_css_selector('div.credit_summary_item')
        director = credits[0].find_element_by_css_selector('a').get_attribute('innerHTML')
        writers =credits[1].find_elements_by_css_selector('a')
    except:
        pass
    if len(writers):
        writer1 = writers[0].get_attribute('innerHTML')
        writers.remove(writers[0])
    else:
        writer1 =float('nan')
    if len(writers):
        writer2 = writers[0].get_attribute('innerHTML')
        writers.remove(writers[0])
    else:
        writer2 =float('nan')
    if len(writers):
        writer3 = writers[0].get_attribute('innerHTML')
        writers.remove(writers[0])
    else:
        writer3 =float('nan')
    #print(writer1,writer2,writer3)

    #Getting Starts of a movie
    stars = credits[2].find_elements_by_css_selector('a')
    stars=stars[0:len(stars)-1]
    if len(stars):
        star1 = stars[0].get_attribute('innerHTML')
        stars.remove(stars[0])
    else:
        star1 =float('nan')
    if len(stars):
        star2 = stars[0].get_attribute('innerHTML')
        stars.remove(stars[0])
    else:
        star2 =float('nan')
    if len(stars):
        star3 = stars[0].get_attribute('innerHTML')
        stars.remove(stars[0])
    else:
        star3 =float('nan')
    if len(stars):
        star4 = stars[0].get_attribute('innerHTML')
        stars.remove(stars[0])
    else:
        star4 =float('nan')

    if len(stars):
        star5 = stars[0].get_attribute('innerHTML')
        stars.remove(stars[0])
    else:
        star5 =float('nan')
    #Getting the Box office details of  a movie
    try:
        content_wide = parent.find_element_by_css_selector('div.flatland').find_element_by_css_selector('div[id="titleStoryLine"]').find_element_by_css_selector('div.see-more').find_elements_by_css_selector('a')
        keywords_list=""
        for ken in content_wide[0:len(content_wide)-2]:
            if len(keywords_list):
                keywords_list = keywords_list+','+ken.find_element_by_css_selector('span[class="itemprop"]').get_attribute('innerHTML')
            #print(ken.get_attribute('innerHTML'))
            else:
                keywords_list = keywords_list+ken.find_element_by_css_selector('span[class="itemprop"]').get_attribute('innerHTML')
    except:
        keywords_list=float('nan')
    
    box_office =parent.find_element_by_css_selector('div.flatland').find_element_by_css_selector('div[id="titleDetails"]').get_attribute('innerHTML')
    bs_obj = bs4.BeautifulSoup(box_office,'html.parser')
    
    box =bs_obj.findAll('div',{'class':'txt-block'})
    Budget,cumm_gross,gross_USA,production_company=float('nan'),float('nan'),float('nan'),float('nan')
    #lst=['January','February','March','April','May','June','July','August','September','November','December']
    for b in box:
        #print(1)
        b=b.text.replace('\n','')
        #print(b)
        if b.startswith('Budget:'):
            Budget= int("".join([ch for ch in b if ch.isdigit()]))
        if b.startswith('Gross USA:'):
            try:
                int(b.split(',')[-1])
                gross_USA= int("".join([ch for ch in b if ch.isdigit()]))
            except:
                b1=b.split(',')[:-1]
                b1 = "".join([ch for ch in b1])
                gross_USA = int("".join([ch for ch in b1 if ch.isdigit()]))
        if b.startswith('Cumulative Worldwide Gross:'):
            try:
                int(b.split(',')[-1])
                cumm_gross = int("".join([ch for ch in b if ch.isdigit()])) 
            except:
                b2=b.split(',')[:-1]
                b2 = "".join([ch for ch in b2])
                cumm_gross = int("".join([ch for ch in b2 if ch.isdigit()])) 
        if b.startswith('Production Co:'):
            production_company =b.replace('Production Co:','').replace('See more »','')
    
    print(gross_USA,cumm_gross)
    imdb_top250=imdb_top250.append({'movie_name':mov_name,
                                   'year':year,
                                   'IMD_Rating':IMD_Rating,
                                   'No_of_reviewers':No_of_reviewers,
                                   'censor_board_rating':censor_board_rating,
                                   'run_time':run_time,
                                   'genre1':genre1,
                                   'genre2':genre2,
                                   'genre3':genre3,
                                   'genre4':genre4,
                                   'release_date':release_date,
                                   'summary':summary,
                                   'writer1':writer1,
                                   'writer2':writer2,
                                   'writer3':writer3,
                                   'star1':star1,
                                   'star2':star2,
                                   'star3':star3,
                                   'star4':star4,
                                   'star5':star5,
                                   'keywords_list':keywords_list,
                                   'Budget':Budget,
                                   'cumm_gross':cumm_gross,
                                   'gross_USA':gross_USA,
                                   'production_company':production_company},ignore_index=True)

imdb_top250

    
```
